```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

from saem import CSEMData
import numpy as np
from scipy.stats import linregress
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from collections import defaultdict

def find_line_intersections(data, main_flight_lines, tie_lines):
    """
    Find intersections between main flight lines and tie lines and return their indices.

    Parameters:
    - data: CSEMData object containing receiver positions and line info.
    - main_flight_lines: List of main flight lines.
    - tie_lines: List of tie lines.

    Returns:
    - intersections: List of intersection receiver index pairs as tuples (tie_line_idx, main_line_idx).
    """
    rx_positions = np.array(data.rx)
    ry_positions = np.array(data.ry)
    line_numbers = np.array(data.line)

    def get_line_params(indices):
        """
        Calculate slope and intercept for a line given indices of points.
        """
        x = rx_positions[indices]
        y = ry_positions[indices]
        slope, intercept, _, _, _ = linregress(x, y)
        return slope, intercept, indices

    # Extract line parameters for main flight lines and tie lines
    main_lines_params = {}
    tie_lines_params = {}
    for line in main_flight_lines:
        indices = np.where(line_numbers == line)[0]
        main_lines_params[line] = get_line_params(indices)
    for line in tie_lines:
        indices = np.where(line_numbers == line)[0]
        tie_lines_params[line] = get_line_params(indices)

    intersections = []
    for main_line, (m1, c1, main_indices) in main_lines_params.items():
        for tie_line, (m2, c2, tie_indices) in tie_lines_params.items():
            if m1 != m2:  # Check if lines are not parallel
                # Calculate intersection point
                x_intersect = (c2 - c1) / (m1 - m2)
                y_intersect = m1 * x_intersect + c1

                # Find the closest points in both lines
                main_dist = np.sqrt((rx_positions[main_indices] - x_intersect) ** 2 +
                                    (ry_positions[main_indices] - y_intersect) ** 2)
                tie_dist = np.sqrt((rx_positions[tie_indices] - x_intersect) ** 2 +
                                   (ry_positions[tie_indices] - y_intersect) ** 2)

                main_closest_idx = main_indices[np.argmin(main_dist)]
                tie_closest_idx = tie_indices[np.argmin(tie_dist)]

                # Add the intersection as a pair of indices
                intersections.append((tie_closest_idx, main_closest_idx))

    return intersections


# Main loop for each transmitter
TxNumber = "Tx1"
path = f"{TxNumber}/"
txpos = path + f"{TxNumber}.kml"
data = CSEMData(datafile=path + "tfN4/*.mat", txPos=txpos)

# Extract positions and lines
rx_positions = np.array(data.rx)
ry_positions = np.array(data.ry)
line_numbers = np.array(data.line)

# Group points by unique lines and calculate slopes
unique_lines = np.unique(line_numbers)
slopes = {}
for line in unique_lines:
    indices = np.where(line_numbers == line)
    x_line = rx_positions[indices]
    y_line = ry_positions[indices]
    slope, _, _, _, _ = linregress(x_line, y_line)
    slopes[line] = slope

# Cluster slopes to identify main flight lines
slope_array = np.array(list(slopes.values())).reshape(-1, 1)
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(slope_array)

# Determine main flight lines and tie lines
centroids = kmeans.cluster_centers_
main_flight_lines = []
tie_lines = []

for line, slope in slopes.items():
    diff_to_centroid_1 = abs(slope - centroids[0])
    diff_to_centroid_2 = abs(slope - centroids[1])
    if diff_to_centroid_1 < diff_to_centroid_2:
        main_flight_lines.append(line)
    else:
        tie_lines.append(line)

# If necessary, swap groups to ensure main_flight_lines has more points
if len(tie_lines) > len(main_flight_lines):
    main_flight_lines, tie_lines = tie_lines, main_flight_lines

# Find intersections between main flight lines and tie lines
intersections = find_line_intersections(data, main_flight_lines, tie_lines)

# Print intersection pairs as (tie_line_idx, main_line_idx)
print("Intersections (tie line index, main line index):")
for tie_idx, main_idx in intersections:
    print(f"({tie_idx}, {main_idx})")

# Plot all data, main flight lines, tie lines, and intersections
plt.figure()
plt.plot(rx_positions, ry_positions, 'o')
for line in main_flight_lines:
    indices = np.where(line_numbers == line)
    plt.plot(rx_positions[indices], ry_positions[indices], color='green')
for line in tie_lines:
    indices = np.where(line_numbers == line)
    plt.plot(rx_positions[indices], ry_positions[indices] , color='red')

# Plot intersections
for tie_idx, main_idx in intersections:
    plt.plot([rx_positions[tie_idx]], [ry_positions[tie_idx]], 'x', color='blue')

plt.title(f"Main and Tie Line Intersections ({TxNumber})")
plt.xlabel("RX Position")
plt.ylabel("RY Position")
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.show()

# Group the intersections by tie line
grouped_by_tie_line = defaultdict(list)

# Populate the dictionary
for tie_idx, main_idx in intersections:
    tie_line = data.line[tie_idx]  # Determine the tie line number from the receiver index
    grouped_by_tie_line[tie_line].append((tie_idx, main_idx))

# Print the grouped pairs
print("Grouped intersections by tie line:")
for tie_line, pairs in grouped_by_tie_line.items():
    print(f"Tie Line {tie_line}: {pairs}")

############################################################################################################################################
# Select the tie line and intersection points
selected_tie_line = list(grouped_by_tie_line.keys())[0]  # Change the index to select a different tie line

# Specify the desired frequency (in Hz)
desired_frequency = 256  # Replace with the desired frequency

# Find the index of the closest frequency in data.f
frequency_index = np.argmin(np.abs(data.f - desired_frequency))
#############################################################################################################################################

# %% line data
# Define the amphi toggle
amphi = True  # Set to True for amplitude and phase, False for real and imaginary
amphi = False
# Define the cmp parameter and dynamically determine the component index and label
cmp = [0, 0, 1]  # Change this to [1, 0, 0] for X or [0, 1, 0] for Y
component_index = cmp.index(1)  # Find the index of the active component
component_label = ['x', 'y', 'z'][component_index]  # Map to 'x', 'y', or 'z'

# Plot the frequency data for the selected tie line
fig, ax = data.showLineFreq(
    line=selected_tie_line,  # Tie line to display
    nf=frequency_index,      # Frequency index found dynamically
    cmp=cmp,                 # Dynamic component selection
    amphi=amphi,             # Toggle between amphi=True and False
    log=True                # Linear scale
)
# X-axis limit based on tie line data
tie_indices = np.where(data.line == selected_tie_line)[0]
x_min = np.min(rx_positions[tie_indices])
x_max = np.max(rx_positions[tie_indices])

for sub_ax in ax.flat:
    sub_ax.set_xlim([x_min, x_max])
    sub_ax.legend()

# Flags to ensure labels are added only once
tie_label_added = False
main_label_added = False

# Overlay intersection points for both tie line and main line
for i, (tie_idx, main_idx) in enumerate(grouped_by_tie_line[selected_tie_line]):
    intersection_x_tie = rx_positions[tie_idx]
    intersection_x_main = rx_positions[main_idx]

    if amphi:
        # Amplitude and phase mode
        tie_amplitude = np.abs(data.DATA[component_index, frequency_index, tie_idx])
        tie_phase = np.angle(data.DATA[component_index, frequency_index, tie_idx], deg=True)

        main_amplitude = np.abs(data.DATA[component_index, frequency_index, main_idx])
        main_phase = np.angle(data.DATA[component_index, frequency_index, main_idx], deg=True)

        # Calculate misfit
        amplitude_misfit = np.abs(tie_amplitude - main_amplitude)*1000
        phase_misfit = np.abs(tie_phase - main_phase)

        # Debug: Print values to verify correctness
        print(f"Intersection {i+1}: Tie Amplitude = {tie_amplitude}, Main Amplitude = {main_amplitude}")
        print(f"Intersection {i+1}: Tie Phase = {tie_phase}, Main Phase = {main_phase}")
        print(f"Intersection {i+1}: Amplitude Misfit = {amplitude_misfit}, Phase Misfit = {phase_misfit}")

        # Plot tie line and main line points
        ax[0, 0].plot(intersection_x_tie, tie_amplitude, 'ro', label='Tie Line' if not tie_label_added else "")
        ax[0, 0].plot(intersection_x_main, main_amplitude, 'go', label='Main Line' if not main_label_added else "")
        ax[0, 0].annotate(f"{amplitude_misfit:.2f}", (intersection_x_tie, tie_amplitude),
                          textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8, color='red')

        ax[1, 0].plot(intersection_x_tie, tie_phase, 'ro', label='Tie Line' if not tie_label_added else "")
        ax[1, 0].plot(intersection_x_main, main_phase, 'go', label='Main Line' if not main_label_added else "")
        ax[1, 0].annotate(f"{phase_misfit:.2f}", (intersection_x_tie, tie_phase),
                          textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8, color='red')
    else:
        # Real and imaginary mode
        tie_real = np.real(data.DATA[component_index, frequency_index, tie_idx])
        tie_imag = np.imag(data.DATA[component_index, frequency_index, tie_idx])

        main_real = np.real(data.DATA[component_index, frequency_index, main_idx])
        main_imag = np.imag(data.DATA[component_index, frequency_index, main_idx])

        # Calculate misfit
        real_misfit = np.abs(tie_real - main_real)*1000
        imag_misfit = np.abs(tie_imag - main_imag)*1000

        # Debug: Print values to verify correctness
        print(f"Intersection {i+1}: Tie Real = {tie_real}, Main Real = {main_real}")
        print(f"Intersection {i+1}: Tie Imag = {tie_imag}, Main Imag = {main_imag}")
        print(f"Intersection {i+1}: Real Misfit = {real_misfit}, Imaginary Misfit = {imag_misfit}")

        # Plot tie line and main line points
        ax[0, 0].plot(intersection_x_tie, tie_real, 'ro', label='Tie Line' if not tie_label_added else "")
        ax[0, 0].plot(intersection_x_main, main_real, 'go', label='Main Line' if not main_label_added else "")
        ax[0, 0].annotate(f"{real_misfit:.2f}", (intersection_x_tie, tie_real),
                          textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8, color='red')

        ax[1, 0].plot(intersection_x_tie, tie_imag, 'ro', label='Tie Line' if not tie_label_added else "")
        ax[1, 0].plot(intersection_x_main, main_imag, 'go', label='Main Line' if not main_label_added else "")
        ax[1, 0].annotate(f"{imag_misfit:.2f}", (intersection_x_tie, tie_imag),
                          textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8, color='red')

    # Update flags to ensure labels are added only once
    tie_label_added = True
    main_label_added = True

# Add legends
for sub_ax in ax.flat:
    sub_ax.legend()

# Show the plot
plt.show()

# Print frequency confirmation
print(f"Plotted data for the frequency closest to {desired_frequency} Hz: {data.f[frequency_index]:.2f} Hz")
# %% sounding
for tie_idx, main_idx in grouped_by_tie_line.get(selected_tie_line, []):
    try:
        # Show sounding for the tie line receiver index
        print(f"Displaying sounding for tie line receiver index {tie_idx}")
        ax = data.showSounding(
            nrx=tie_idx,
            cmp=cmp,
            label=f'B{component_label} tie',
            color='blue'
        )
        
        # Validate that ax is correctly initialized
        if ax is None:
            raise ValueError("Failed to initialize plot axis (ax) for tie line receiver.")
        
        # Show sounding for the main line receiver index
        print(f"Displaying sounding for main line receiver index {main_idx}")
        data.showSounding(
            nrx=main_idx,
            cmp=cmp,
            ax=ax,
            label=f'B{component_label} main',
            color='green'
        )
        
    except KeyError as e:
        print(f"KeyError: {e} - Ensure 'selected_tie_line' exists in 'grouped_by_tie_line'")
    except ValueError as e:
        print(f"ValueError: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```
